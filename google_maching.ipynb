```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.drop(columns='type'),df['type'],random_state=42,test_size=0.3)
```

```python
x_train.shape,x_test.shape
```

```python
from sklearn.compose import ColumnTransformer
```

```python
transform=ColumnTransformer(transformers=[
    ('t1',OneHotEncoder(sparse_output=False,drop='first'),['name','brand','category','country','device','Variant2'])
],remainder='passthrough')
```

```python
transform.fit(x_train)
```

```python
x_train_trainsform=transform.transform(x_train).astype(int)
x_test_trainsform=transform.transform(x_test).astype(int)
```

```python
x_train_trainsform=pd.DataFrame(x_train_trainsform,columns=transform.get_feature_names_out())
```

```python
x_test_trainsform=pd.DataFrame(x_test_trainsform,columns=transform.get_feature_names_out())
```

```python
x_train_trainsform.to_csv('x_train_trainsform',index=False)
x_test_trainsform.to_csv('x_test_trainsform',index=False)
y_train.to_csv('y_train',index=False)
y_test.to_csv('y_test',index=False)
```
